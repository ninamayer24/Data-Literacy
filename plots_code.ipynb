```python
df = data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

# Optional: for interactive web maps if lon/lat are present
try:
    import folium
    HAS_FOLIUM = True
except Exception:
    HAS_FOLIUM = False

# Helper to safely cast numerics
def to_num(s):
    return pd.to_numeric(s, errors='coerce')

# Cast important numeric columns
num_cols = [
    'Baujahr Überbau','Baujahr Unterbau','Zustandsnote','Traglastindex',
    'Länge (m)','Breite (m)','Fläche (qm)','Substanzkennzahl'
]
for c in num_cols:
    if c in df.columns:
        df[c] = to_num(df[c])

# Trim common string columns
for c in ['Ort','Kreis','Bundesland','Bundeslandname','Baustoff Überbau','Baustoffklasse',
          'Konstruktion','Sperrung SV','Zugeordneter Sachverhalt','Zustandsnotenklasse']:
    if c in df.columns:
        df[c] = df[c].astype('string').str.strip()

# Derive a unified "construction year" (prefer Überbau, fallback Unterbau)
df['Baujahr'] = df['Baujahr Überbau'].fillna(df['Baujahr Unterbau'])
df['Baujahr'] = df['Baujahr'].astype(float)

# Build decade bins (e.g., 1970, 1980, ...)
def year_to_decade(y):
    if pd.isna(y):
        return np.nan
    return int(y // 10 * 10)

df['Decade'] = df['Baujahr'].apply(year_to_decade)



# 4) Distribution of Zustandsnote per Bundesland


if 'Zustandsnote' in df.columns and 'Bundeslandname' in df.columns:
    zdata = df[['Bundeslandname','Zustandsnote']].dropna()
    # Boxplot: Zustandsnote by Bundesland (only Bundesländer with enough data)
    counts = zdata['Bundeslandname'].value_counts()
    keep_states = counts[counts >= 30].index  # drop very small groups for readability
    zdata_f = zdata[zdata['Bundeslandname'].isin(keep_states)]

    # Prepare data for matplotlib boxplot
    order = zdata_f.groupby('Bundeslandname')['Zustandsnote'].median().sort_values().index
    grouped = [zdata_f.loc[zdata_f['Bundeslandname'] == bl, 'Zustandsnote'].values for bl in order]

    plt.figure(figsize=(12, 6))
    plt.boxplot(grouped, showfliers=False)
    plt.xticks(range(1, len(order)+1), order, rotation=45, ha='right')
    plt.xlabel('Bundesland'); plt.ylabel('Zustandsnote')
    plt.title('Distribution of bridge condition (Zustandsnote) by Bundesland')
    plt.tight_layout()
    plt.show()




# 6) Correlation between construction year and material usage (with plot)


material_col = 'Baustoff Überbau' if 'Baustoff Überbau' in df.columns else (
               'Baustoffklasse' if 'Baustoffklasse' in df.columns else None)

if material_col is not None:
    mat = df[['Decade', material_col]].dropna()
    # Limit to sensible decades (>= 1850) for readability
    mat = mat[(mat['Decade'] >= 1850) & (mat['Decade'] <= pd.Timestamp.now().year)]
    # Build shares by decade
    ct = pd.crosstab(mat['Decade'], mat[material_col])
    shares = ct.div(ct.sum(axis=1), axis=0)

    # Plot as stacked area (single chart; multiple series in one figure)
    plt.figure(figsize=(12, 6))
    x = shares.index.values
    y = shares.values.T  # shape: materials x decades
    plt.stackplot(x, y, labels=shares.columns)  # no explicit colors per requirement
    plt.legend(loc='upper left', ncol=2, fontsize=8)
    plt.xlabel('Decade'); plt.ylabel('Share')
    plt.title('Material usage share by decade (Überbau)')
    plt.tight_layout()
    plt.show()

    # Optional: chi-squared association between decade and material
    try:
        from scipy.stats import chi2_contingency
        chi2, p, dof, _ = chi2_contingency(ct)
        print(f"Chi²(Decade vs Material) = {chi2:.1f}, dof={dof}, p-value={p:.3g}")
    except Exception:
        pass
else:
    print("No material column found for material-vs-year analysis.")

# 7) Development of construction years over decades


dev = df['Decade'].dropna().astype(int)
dev = dev[(dev >= 1850) & (dev <= pd.Timestamp.now().year)]
decade_counts = dev.value_counts().sort_index()

plt.figure(figsize=(10, 5))
plt.bar(decade_counts.index.astype(int), decade_counts.values, width=8)
plt.xlabel('Decade'); plt.ylabel('Number of bridges')
plt.title('Number of bridges built per decade')
plt.tight_layout()
plt.show()

print("\nDone.")
```

```python
# -*- coding: utf-8 -*-
# Correlations vs. "Zustandsnote" + standalone distribution plots
# Comments in English as requested.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path


# Helpers
def to_num(s):
    return pd.to_numeric(s, errors='coerce')

# Cast numerics we will use
num_cols = [
    'Zustandsnote','Baujahr Überbau','Baujahr Unterbau','Traglastindex',
    'Länge (m)','Breite (m)','Fläche (qm)','Substanzkennzahl'
]
for c in num_cols:
    if c in df.columns:
        df[c] = to_num(df[c])

# Unify construction year and derive age
df['Baujahr'] = df['Baujahr Überbau'].fillna(df.get('Baujahr Unterbau'))
df['Baujahr'] = to_num(df['Baujahr'])
current_year = pd.Timestamp.now().year

# Common categorical columns (strip)
for c in ['Baustoff Überbau','Baustoffklasse','Konstruktion','Bundeslandname','Kreis']:
    if c in df.columns:
        df[c] = df[c].astype('string').str.strip()


# 1) Standalone distribution of "Zustandsnote"

zcol = 'Zustandsnote'
if zcol in df.columns:
    z = df[zcol].dropna()



    # Bar counts (discrete view if notes are on a small set like 1.0..4.0)
    counts = z.round(1).value_counts().sort_index()
    plt.figure(figsize=(8, 5))
    plt.bar(counts.index.astype(str), counts.values)
    plt.xlabel('Zustandsnote')
    plt.ylabel('Count')
    plt.title('Zustandsnote frequency')
    plt.tight_layout()
    plt.show()
else:
    print("Column 'Zustandsnote' not found — aborting plots.")


# Utility: numeric scatter with linear fit & Spearman rho

def scatter_with_fit_and_spearman(x, y, xlab, ylab, title):
    """Matplotlib scatter; add linear fit; show Spearman rho in title."""
    from scipy.stats import spearmanr

    d = pd.DataFrame({'x': x, 'y': y}).dropna()
    if d.empty or d['x'].nunique() < 2 or d['y'].nunique() < 2:
        print(f"Not enough data for {title}")
        return

    # Spearman correlation (robust to non-linear monotonic relationships)
    rho, p = spearmanr(d['x'], d['y'], nan_policy='omit')

    # Linear fit (least squares)
    m, b = np.polyfit(d['x'], d['y'], 1)
    xline = np.linspace(d['x'].min(), d['x'].max(), 100)
    yline = m * xline + b

    plt.figure(figsize=(7, 5))
    plt.scatter(d['x'], d['y'], s=10, alpha=0.6)
    #plt.plot(xline, yline)
    plt.xlabel(xlab); plt.ylabel(ylab)
    plt.title(f"{title}\nSpearman ρ = {rho:.3f} (p={p:.3g})")
    plt.tight_layout()
    plt.show()

# -----------------------------
# 2) Numeric correlations vs. Zustandsnote
# -----------------------------
if zcol in df.columns:
    numeric_targets = [
        ('Baujahr', 'Year (Baujahr)'),
        ('Traglastindex', 'Load index'),
        ('Länge (m)', 'Length (m)'),
        ('Breite (m)', 'Width (m)'),
        ('Fläche (qm)', 'Area (sqm)'),
        ('Substanzkennzahl', 'Substance index')
    ]
    for col, lab in numeric_targets:
        if col in df.columns:
            scatter_with_fit_and_spearman(df[col], df[zcol], lab, 'Zustandsnote',
                                          f'{lab} vs. Zustandsnote')


# 3) Spearman overview: Zustandsnote vs. ALL numeric features

if zcol in df.columns:
    # pick all numeric cols
    numeric_df = df.select_dtypes(include=[np.number]).copy()
    if zcol in numeric_df.columns:
        # Compute Spearman vs all others
        from scipy.stats import spearmanr
        rho_list = []
        for c in numeric_df.columns:
            if c == zcol:
                continue
            sub = numeric_df[[c, zcol]].dropna()
            if sub[c].nunique() < 2 or sub[zcol].nunique() < 2:
                continue
            r, p = spearmanr(sub[c], sub[zcol], nan_policy='omit')
            rho_list.append((c, r, p, len(sub)))

        if rho_list:
            corr_df = pd.DataFrame(rho_list, columns=['Feature','Spearman_rho','p_value','N']).sort_values('Spearman_rho')
            print("\nSpearman correlation with Zustandsnote (sorted):")
            print(corr_df.to_string(index=False, float_format=lambda v: f"{v:.4f}" if isinstance(v, float) else str(v)))

            # Simple bar plot of Spearman rho (top 15 by |rho|)
            corr_df['abs_rho'] = corr_df['Spearman_rho'].abs()
            top = corr_df.sort_values('abs_rho', ascending=False).head(15)

            plt.figure(figsize=(10, 6))
            plt.bar(top['Feature'].astype(str), top['Spearman_rho'])
            plt.xticks(rotation=45, ha='right')
            plt.ylabel('Spearman rho')
            plt.title('Top correlations with Zustandsnote (Spearman)')
            plt.tight_layout()
            plt.show()


# 4) Categorical vs. Zustandsnote (Boxplots + Kruskal-Wallis)

def boxplot_by_category(frame, cat_col, value_col, min_count=30):
    """Boxplot of value_col by category; drops small groups; Kruskal-Wallis test."""
    from scipy.stats import kruskal

    d = frame[[cat_col, value_col]].dropna()
    if d.empty:
        print(f"No data for {cat_col} vs {value_col}")
        return

    counts = d[cat_col].value_counts()
    keep = counts[counts >= min_count].index
    d = d[d[cat_col].isin(keep)]
    if d[cat_col].nunique() < 2:
        print(f"Not enough categories with >= {min_count} samples for {cat_col}")
        return

    # Prepare groups for Kruskal test and boxplot
    order = d.groupby(cat_col)[value_col].median().sort_values().index
    groups = [d.loc[d[cat_col] == k, value_col].values for k in order]
    try:
        stat, p = kruskal(*groups)
        stat_line = f"Kruskal-Wallis H = {stat:.2f}, p = {p:.3g}"
    except Exception:
        stat_line = "Kruskal-Wallis not computed."

    plt.figure(figsize=(12, 6))
    plt.boxplot(groups, showfliers=False)
    plt.xticks(range(1, len(order)+1), order, rotation=45, ha='right')
    plt.xlabel(cat_col)
    plt.ylabel(value_col)
    plt.title(f'{value_col} by {cat_col}\n{stat_line}')
    plt.tight_layout()
    plt.show()



print("\nDone: distribution + numeric/categorical correlations vs. Zustandsnote.")
```
