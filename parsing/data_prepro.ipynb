```python
# load libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# read data
data = pd.read_csv('Brückenstatistik_Deutschland_-575782646687357632.csv', sep=';')
#data.head()

# exploratory data analysis
print(f"There are {data.shape[0]} rows and {data.shape[1]} columns in the dataset.")
print("Column names:", data.columns.tolist())
bridges_in_tuebingen = [1 if x == 'Tübingen' else 0 for x in data['Ort']]
print(f"There are {sum(bridges_in_tuebingen)} bridges in Tübingen.")
print(f"Oldest bridge with respect to the substructure built in: {data['Baujahr Unterbau'].min()}")
```

```python
# plot histogram of bridge ages
n, bins, patches = plt.hist(data['Baujahr Überbau'], edgecolor='black', range=(1000, 2025), bins=25)
plt.title('Distribution of Bridge (superstructure) Ages in Germany')
plt.xlabel('Year of Construction')
plt.ylabel('Number of Bridges')
plt.show()

#print(f"Number of bridges in each bin: {n}")

# print every year and how many bridges were built that year (for superstructure)
year_counts = data['Baujahr Unterbau'].value_counts().sort_index()
for year, count in year_counts.items():
    if year == 0:
        continue  # skip year 0 entries
    print(f"Year: {int(year)}, Number of bridges built: {count}")  
```
