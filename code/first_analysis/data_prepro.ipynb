# Analyse: Zusammenhang zwischen Zustandsnote und Baujahr
Dieser Abschnitt bereinigt die Daten, erstellt Kennzahlen und Visualisierungen und führt einfache statistische Tests durch, um Zusammenhänge zwischen `Zustandsnote` und `Baujahr` (bzw. Alter) zu untersuchen.
- Annahmen: Spaltennamen `Zustandsnote`, `Baujahr Unterbau`, `Baujahr Überbau` sollten existieren.
- Falls andere Spaltennamen verwendet werden, passe die Namen im nachfolgenden Code an.

```python
#%% Load packages (install with pip if needed)
# %pip install pandas numpy matplotlib seaborn scipy statsmodels
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import statsmodels.api as sm
import statsmodels.formula.api as smf
from datetime import datetime

# --- Load data ---
fn = 'Brückenstatistik_Deutschland_-575782646687357632.csv'
data = pd.read_csv(fn, sep=';', encoding='utf-8', low_memory=False)

# --- Normalize column names ---
data.columns = [c.strip() for c in data.columns]
print(f'Read {data.shape[0]} rows and {data.shape[1]} columns')

# --- Helper to clean year fields ---
def clean_year(x):
    try:
        y = pd.to_numeric(x, errors='coerce')
        if pd.isna(y) or y == 0 or y < 1000:
            return np.nan
        return int(y)
    except:
        return np.nan

# create standardized baujahr columns (if present)
if 'Baujahr Überbau' in data.columns:
    data['baujahr_ueberbau'] = data['Baujahr Überbau'].apply(clean_year)
else:
    data['baujahr_ueberbau'] = np.nan
if 'Baujahr Unterbau' in data.columns:
    data['baujahr_unterbau'] = data['Baujahr Unterbau'].apply(clean_year)
else:
    data['baujahr_unterbau'] = np.nan

# prefer superstructure year (Überbau) if available, else Unterbau
data['baujahr'] = data['baujahr_ueberbau'].combine_first(data['baujahr_unterbau'])

# compute age relative to current year
current_year = datetime.now().year
data['age'] = current_year - data['baujahr']

# Clean Zustandsnote into numeric column
if 'Zustandsnote' in data.columns:
    data['Zustandsnote_clean'] = data['Zustandsnote'].astype(str).str.replace(',', '.').replace({'nan':'', 'None':''})
    data['Zustandsnote_num'] = pd.to_numeric(data['Zustandsnote_clean'], errors='coerce')
else:
    data['Zustandsnote_num'] = np.nan

# Subset to rows with both values present
df = data.loc[data['baujahr'].notna() & data['Zustandsnote_num'].notna()].copy()
print(f'Rows with both Baujahr and Zustandsnote: {df.shape[0]}')

# --- Descriptive statistics ---
display(df[['baujahr','age','Zustandsnote_num']].describe())

# --- Plots ---
sns.set(style='whitegrid', context='talk')

# Histogram of Baujahr and Age
fig, axes = plt.subplots(1,2, figsize=(16,5))
sns.histplot(df['baujahr'].dropna(), bins=30, kde=False, ax=axes[0])
axes[0].set_title('Baujahr')
axes[0].set_xlabel('Year')
sns.histplot(df['age'].dropna(), bins=30, kde=False, ax=axes[1])
axes[1].set_title('Age (years)')
axes[1].set_xlabel('Age')
plt.tight_layout()
plt.show()

# Boxplot of Zustandsnote by decade
df['decade'] = (df['baujahr'] // 10) * 10
plt.figure(figsize=(12,6))
sns.boxplot(x='decade', y='Zustandsnote_num', data=df, showfliers=False)
plt.xticks(rotation=45)
plt.xlabel('Decade')
plt.ylabel('Zustandsnote')
plt.title('Zustandsnote distribution by construction decade')
plt.show()

# Scatter age vs Zustandsnote with LOWESS
plt.figure(figsize=(10,6))
sns.regplot(x='age', y='Zustandsnote_num', data=df, scatter_kws={'alpha':0.4,'s':10}, lowess=True)
plt.xlabel('Age (years)')
plt.ylabel('Zustandsnote')
plt.title('Age vs Zustandsnote (LOWESS)')
plt.show()

# Violin + swarm of age per rounded Zustandsnote
df['zust_round'] = df['Zustandsnote_num'].round(0).astype('Int64')
plt.figure(figsize=(10,6))
sns.violinplot(x='zust_round', y='age', data=df, inner=None, color='.8')
sns.stripplot(x='zust_round', y='age', data=df, jitter=0.2, size=3, alpha=0.6)
plt.xlabel('Rounded Zustandsnote')
plt.ylabel('Age (years)')
plt.title('Age distribution by rounded Zustandsnote')
plt.show()

```

```python
# Regionale Vergleiche: Top 3 / Flop 3 für 'Bundeslandname' und 'Kreis'
# Benötigte Pakete: pandas, numpy, matplotlib, seaborn, scipy
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from IPython.display import display

sns.set(style="whitegrid", context="talk")

# Erstelle Ordner für Ausgaben
os.makedirs("figures", exist_ok=True)
os.makedirs("summaries", exist_ok=True)

# Falls 'df' nicht vorhanden ist: sichere Vorverarbeitung (verwendet dieselbe Logik wie vorher)
def prepare_df(filepath='Brückenstatistik_Deutschland_-575782646687357632.csv'):
    d = pd.read_csv(filepath, sep=';', encoding='utf-8', low_memory=False)
    d.columns = [c.strip() for c in d.columns]
    def clean_year(x):
        try:
            y = pd.to_numeric(x, errors='coerce')
            if pd.isna(y) or y == 0 or y < 1000:
                return np.nan
            return int(y)
        except:
            return np.nan
    if 'Baujahr Überbau' in d.columns:
        d['baujahr_ueberbau'] = d['Baujahr Überbau'].apply(clean_year)
    else:
        d['baujahr_ueberbau'] = np.nan
    if 'Baujahr Unterbau' in d.columns:
        d['baujahr_unterbau'] = d['Baujahr Unterbau'].apply(clean_year)
    else:
        d['baujahr_unterbau'] = np.nan
    d['baujahr'] = d['baujahr_ueberbau'].combine_first(d['baujahr_unterbau'])
    from datetime import datetime
    current_year = datetime.now().year
    d['age'] = current_year - d['baujahr']
    if 'Zustandsnote' in d.columns:
        d['Zustandsnote_clean'] = d['Zustandsnote'].astype(str).str.replace(',', '.').replace({'nan':'', 'None':''})
        d['Zustandsnote_num'] = pd.to_numeric(d['Zustandsnote_clean'], errors='coerce')
    else:
        d['Zustandsnote_num'] = np.nan
    return d

if 'df' not in globals():
    print("Kein vorhandenes 'df' gefunden — bereite DataFrame jetzt vor.")
    data = prepare_df()
    df = data.loc[data['baujahr'].notna() & data['Zustandsnote_num'].notna()].copy()
else:
    print("Verwende vorhandenes DataFrame 'df' aus vorheriger Zelle.")

print(f"Analysiere {df.shape[0]} Zeilen mit Baujahr und Zustandsnote.")

def compare_regions(df, region_col, top_n=3, min_count=20, save_prefix='figures/regions'):
    """
    Top / Flop Regionen vergleichen (region_col z.B. 'Bundeslandname' oder 'Kreis').
    Speichert Summary CSV in 'summaries/' und Plots in 'figures/'.
    """
    if region_col not in df.columns:
        print(f"Spalte '{region_col}' nicht gefunden.")
        return None

    # Aggregation je Region
    summary = df.groupby(region_col).agg(
        count=('Zustandsnote_num','count'),
        mean_zust=('Zustandsnote_num','mean'),
        median_zust=('Zustandsnote_num','median'),
        mean_age=('age','mean')
    ).reset_index().sort_values('count', ascending=False)

    # Speichern
    csv_name = f"summaries/summary_by_{region_col}.csv"
    summary.to_csv(csv_name, index=False)
    print(f"Summary gespeichert: {csv_name}")

    # Auswahl nur Regionen mit mind. min_count Beobachtungen
    sel = summary[summary['count'] >= min_count].copy()
    if sel.empty:
        print(f"Keine Regionen mit >= {min_count} Beobachtungen in '{region_col}'.")
        return summary

    best = sel.nsmallest(top_n, 'mean_zust')
    worst = sel.nlargest(top_n, 'mean_zust')

    print(f"\nTop {top_n} (beste Durchschnitts‑Zustandsnote) in '{region_col}':")
    display(best)
    print(f"\nFlop {top_n} (schlechteste Durchschnitts‑Zustandsnote) in '{region_col}':")
    display(worst)

    # Kombinierte Boxplots (best zuerst, dann flop)
    compare_order = list(best[region_col]) + list(worst[region_col])
    df_cmp = df[df[region_col].isin(compare_order)].copy()

    plt.figure(figsize=(12,6))
    sns.boxplot(x=region_col, y='Zustandsnote_num', data=df_cmp, order=compare_order, showfliers=False)
    sns.stripplot(x=region_col, y='Zustandsnote_num', data=df_cmp, order=compare_order,
                  jitter=0.25, size=3, color='k', alpha=0.5)
    plt.xticks(rotation=45)
    plt.ylabel('Zustandsnote')
    plt.title(f'Zustandsnote: Top{top_n} vs Flop{top_n} - {region_col}')
    fname_box = f"{save_prefix}_{region_col}_top{top_n}_flop{top_n}_boxplot.png"
    plt.tight_layout()
    plt.savefig(fname_box, dpi=150, bbox_inches='tight')
    print(f"Boxplot gespeichert: {fname_box}")
    plt.show()

    # Kruskal-Wallis über Regionen (nur Regionen mit mind. min_count)
    groups = [g['Zustandsnote_num'].values for _, g in df.groupby(region_col) if len(g) >= min_count]
    if len(groups) >= 2:
        kw_stat, kw_p = stats.kruskal(*groups)
        print(f"Kruskal-Wallis über Regionen (≥{min_count} Obs): H={kw_stat:.3f}, p={kw_p:.3e}")
    else:
        print("Nicht genug Regionen für Kruskal-Wallis.")

    # Mann-Whitney U Test: vereinigt Top vs vereinigt Flop
    top_vals = df[df[region_col].isin(best[region_col])]['Zustandsnote_num'].dropna().values
    flop_vals = df[df[region_col].isin(worst[region_col])]['Zustandsnote_num'].dropna().values
    if len(top_vals) >= 5 and len(flop_vals) >= 5:
        u_stat, u_p = stats.mannwhitneyu(top_vals, flop_vals, alternative='two-sided')
        print(f"Mann-Whitney U (Top{top_n} vs Flop{top_n}): U={u_stat:.3f}, p={u_p:.3e}")
    else:
        print("Nicht genug Beobachtungen für Mann-Whitney (jeweils >=5 erforderlich).")

    # Scatter: mittleres Alter vs mittlere Zustandsnote aller Regionen (Speichern)
    plt.figure(figsize=(8,6))
    sns.regplot(x='mean_age', y='mean_zust', data=summary, scatter_kws={'s':40,'alpha':0.7})
    plt.xlabel('Mittleres Alter (Jahre)')
    plt.ylabel('Mittlere Zustandsnote')
    plt.title(f'Mittleres Alter vs Mittlere Zustandsnote nach {region_col}')
    fname_scatter = f"{save_prefix}_{region_col}_mean_age_vs_mean_zust.png"
    plt.tight_layout()
    plt.savefig(fname_scatter, dpi=150, bbox_inches='tight')
    print(f"Scatter gespeichert: {fname_scatter}")
    plt.show()

    return {"summary": summary, "best": best, "worst": worst}

# --- Aufrufe für deine Spalten ---
# Bundeslandname (bessere Lesbarkeit als 'Bundesland')
res_bundesland = compare_regions(df, 'Bundeslandname', top_n=3, min_count=30, save_prefix='figures/regions')

# Kreis (Landkreis-ähnlich): evtl. härtere Mindestgrenze, z.B. min_count=50
res_kreis = compare_regions(df, 'Kreis', top_n=3, min_count=50, save_prefix='figures/regions')

# Optional: zeige zusammenfassende CSV-Dateien
print(
)
```

# Choroplethenkarte: Zustandsnote pro Kreis
Diese Zelle erzeugt eine Deutschlandkarte (Kreise) und färbt die Kreise nach mittlerer `Zustandsnote`.
Vorgehen:
- Versucht zuerst, eine lokale GeoJSON/Shapefile (`landkreise.geojson`, `kreise.geojson`, `germany_counties.geojson`) zu verwenden.
- Falls nicht vorhanden, versucht die Zelle, eine öffentliche GeoJSON herunterzuladen (nur wenn online verfügbar).
- Führt ein Abgleich der Kreisnamen zwischen Ihrem DataFrame `df['Kreis']` und dem GeoJSON durch (exakt, dann fuzzy).
- Speichert die Karte in `figures/choropleth_kreis.png`.

```python
# in einer Notebook-Zelle (oder PowerShell):
%pip install geopandas matplotlib descartes
```

```python
#%% Choroplethenkarte pro Kreis (Kreis)
# Benötigte Pakete: geopandas, matplotlib, descartes (evtl. pip install geopandas)
import os
import json
import difflib
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
from urllib.request import urlopen

os.makedirs('figures', exist_ok=True)

# Helper: try common local filenames for GeoJSON
geo_paths = [
    'landkreise.geojson', 'kreise.geojson', 'germany_counties.geojson',
    'landkreise.shp', 'kreise.shp'
]
geo_path = None
for p in geo_paths:
    if os.path.exists(p):
        geo_path = p
        break

# If not found locally, attempt to download a public GeoJSON (may fail offline)
if geo_path is None:
    print('Keine lokale GeoJSON/Shapefile gefunden. Versuche, eine öffentliche GeoJSON herunterzuladen...')
    # Public source (GitHub raw) - change or allow download if your environment blocks it
    url_candidates = [
        'https://raw.githubusercontent.com/isellsoap/deutschlandGeoJSON/master/landkreise/landkreise.geo.json',
        'https://raw.githubusercontent.com/ginseng666/germany-geojson/master/5km/landkreise.geojson'
    ]
    for url in url_candidates:
        try:
            print('Versuche:', url)
            resp = urlopen(url)
            gj = json.load(resp)
            # save locally for reuse
            with open('landkreise_downloaded.geojson', 'w', encoding='utf-8') as f:
                json.dump(gj, f)
            geo_path = 'landkreise_downloaded.geojson'
            print('GeoJSON heruntergeladen und als', geo_path, 'gespeichert')
            break
        except Exception as e:
            print('Download fehlgeschlagen:', e)

if geo_path is None:
    raise RuntimeError('Keine GeoJSON verfügbar. Bitte lege eine GeoJSON für Landkreise als landkreise.geojson im Arbeitsverzeichnis ab oder erlaube den Download.')

# Load GeoDataFrame
print('Lese GeoJSON/Shapefile:', geo_path)
gdf = gpd.read_file(geo_path)
print('Geo-Spalten:', list(gdf.columns))

# Prepare summary per Kreis (use saved CSV if exists)
summary_csv = 'summaries/summary_by_Kreis.csv'
if os.path.exists(summary_csv):
    summary = pd.read_csv(summary_csv)
else:
    summary = df.groupby('Kreis').agg(count=('Zustandsnote_num','count'), mean_zust=('Zustandsnote_num','mean'), mean_age=('age','mean')).reset_index()

# Identify candidate name columns in GeoDataFrame (common: 'GEN', 'NAME', 'name', 'NUTS_NAME')
name_cols = [c for c in gdf.columns if any(x in c.lower() for x in ['name','gen','gemeinde','county','kreis'])]
print('Mögliche Namensspalten im GeoJSON:', name_cols)

# Choose a name column (prefer 'GEN' or 'name')
if 'GEN' in gdf.columns:
    geo_name_col = 'GEN'
elif 'name' in gdf.columns:
    geo_name_col = 'name'
elif name_cols:
    geo_name_col = name_cols[0]
else:
    geo_name_col = gdf.columns[0]
print('Verwende Geo-Name-Spalte:', geo_name_col)

# Attempt exact merge on name
merged = gdf.merge(summary, left_on=geo_name_col, right_on='Kreis', how='left')
missing = merged['mean_zust'].isna().sum()
print(f'Nach exaktem Merge fehlen Werte für {missing} Geometrien')

# If many missing, try fuzzy matching for unmatched geo names
if missing > 0:
    # build mapping from geo name -> best match in summary['Kreis'] (if above threshold)
    kreise_names = summary['Kreis'].astype(str).unique().tolist()
    mapping = {}
    for gn in gdf[geo_name_col].astype(str):
        if gn in kreise_names:
            mapping[gn] = gn
            continue
        # find close match
        matches = difflib.get_close_matches(gn, kreise_names, n=1, cutoff=0.8)
        if matches:
            mapping[gn] = matches[0]
        else:
            mapping[gn] = None
    # create column with matched Kreis name
    gdf['Kreis_match'] = gdf[geo_name_col].map(mapping)
    merged = gdf.merge(summary, left_on='Kreis_match', right_on='Kreis', how='left')
    missing2 = merged['mean_zust'].isna().sum()
    print(f'Nach fuzzy-match fehlen Werte für {missing2} Geometrien')

# If still missing many, show a sample to let user decide mapping adjustments
if merged['mean_zust'].isna().sum() > 0:
    print('\nBeispiele fehlender Geo-Zuordnungen (geo name -> matched):')
    display(merged.loc[merged['mean_zust'].isna(), [geo_name_col, 'Kreis_match']].head(20))
    print('\nWenn viele fehlen, kannst du die Kreis-Spaltennamen prüfen oder eine manuelle Zuordnungsliste bereitstellen.')

# Plot choropleth (use mean_zust)
fig, ax = plt.subplots(1,1, figsize=(10,12))
merged.plot(column='mean_zust', cmap='RdYlGn_r', linewidth=0.2, ax=ax, edgecolor='0.8', legend=True,
            legend_kwds={'label': 'Mittlere Zustandsnote', 'shrink': 0.6})
ax.axis('off')
ax.set_title('Mittlere Zustandsnote pro Kreis')
outname = 'figures/choropleth_kreis.png'
plt.savefig(outname, dpi=150, bbox_inches='tight')
print('Choropleth gespeichert als', outname)
plt.show()
```

```python
#%% Interaktive Folium‑Karte (Speicherung als HTML)
# Install (uncomment if needed)
# %pip install folium
import os
import json
import difflib
import pandas as pd
import geopandas as gpd
import folium
from folium.features import GeoJsonTooltip
from urllib.request import urlopen

os.makedirs('figures', exist_ok=True)

# Locate or download GeoJSON (re-use logic from previous cell)
geo_candidates = ['landkreise.geojson','kreise.geojson','landkreise_downloaded.geojson','landkreise.geo.json']
geo_path = next((p for p in geo_candidates if os.path.exists(p)), None)
if geo_path is None:
    # attempt download (same candidates as before)
    urls = ['https://raw.githubusercontent.com/isellsoap/deutschlandGeoJSON/master/landkreise/landkreise.geo.json',
            'https://raw.githubusercontent.com/ginseng666/germany-geojson/master/5km/landkreise.geojson']
    for u in urls:
        try:
            print('Versuche Download:', u)
            gj = json.load(urlopen(u))
            with open('landkreise_downloaded.geojson','w', encoding='utf-8') as f:
                json.dump(gj, f)
            geo_path = 'landkreise_downloaded.geojson'
            print('GeoJSON gespeichert als', geo_path)
            break
        except Exception as e:
            print('Download fehlgeschlagen:', e)

if geo_path is None:
    raise RuntimeError('Keine GeoJSON verfügbar — speichere eine Datei namens landkreise.geojson im Arbeitsverzeichnis oder erlaube Download.')

# load and prepare geo dataframe
gdf = gpd.read_file(geo_path)
# ensure lon/lat (WGS84)
if gdf.crs is None:
    try:
        gdf = gdf.to_crs(epsg=4326)
    except Exception:
        pass
else:
    gdf = gdf.to_crs(epsg=4326)

# load summary (or compute)
summary_csv = 'summaries/summary_by_Kreis.csv'
if os.path.exists(summary_csv):
    summary = pd.read_csv(summary_csv)
else:
    summary = df.groupby('Kreis').agg(count=('Zustandsnote_num','count'), mean_zust=('Zustandsnote_num','mean'), mean_age=('age','mean')).reset_index()

# find likely name column in geojson
candidates = [c for c in gdf.columns if any(k in c.lower() for k in ['name','gen','kreis','county'])]
geo_name = 'GEN' if 'GEN' in gdf.columns else ( 'name' if 'name' in gdf.columns else (candidates[0] if candidates else gdf.columns[0]))
print('Geo name column chosen:', geo_name)

# exact merge first
gdf['__geo_name__'] = gdf[geo_name].astype(str)
merged = gdf.merge(summary, left_on='__geo_name__', right_on='Kreis', how='left')
missing = merged['mean_zust'].isna().sum()
print('Nach exaktem Merge fehlende Geometrien:', missing)

# fuzzy if needed
if missing > 0:
    kreise = summary['Kreis'].astype(str).unique().tolist()
    map_match = {}
    for name in gdf['__geo_name__'].astype(str):
        if name in kreise:
            map_match[name] = name
            continue
        m = difflib.get_close_matches(name, kreise, n=1, cutoff=0.78)
        map_match[name] = m[0] if m else None
    gdf['Kreis_match'] = gdf['__geo_name__'].map(map_match)
    merged = gdf.merge(summary, left_on='Kreis_match', right_on='Kreis', how='left')
    missing2 = merged['mean_zust'].isna().sum()
    print('Nach fuzzy matching fehlende Geometrien:', missing2)

# show sample of unmatched for manual mapping if still many missing
if merged['mean_zust'].isna().sum() > 0:
    display(merged.loc[merged['mean_zust'].isna(), ['__geo_name__','Kreis_match']].head(30))
    print('Wenn viele fehlen, erstelle bitte eine manuelle Mapping-Tabelle (GeoName -> Kreis) und ich helfe beim Einbinden.')

# prepare GeoJSON for folium (ensure property 'Kreis' exists for key_on)
# merged already has column 'Kreis' from the join; convert to geojson string
geojson_data = json.loads(merged.to_json())

# build folium map centered on Germany
m = folium.Map(location=[51.1657, 10.4515], zoom_start=6, tiles='cartodbpositron')

# Choropleth layer: use feature.properties.Kreis as key (must match summary 'Kreis')
folium.Choropleth(
    geo_data=geojson_data,
    data=summary,
    columns=['Kreis','mean_zust'],
    key_on='feature.properties.Kreis',
    fill_color='YlOrRd',
    fill_opacity=0.8,
    line_opacity=0.2,
    legend_name='Mittlere Zustandsnote'
).add_to(m)

# Add tooltip with name and mean_zust and count (if present)
tooltip = GeoJsonTooltip(fields=['Kreis','mean_zust','count'], aliases=['Kreis','Mean Zustandsnote','Count'], localize=True)
folium.GeoJson(geojson_data, name='Kreise', tooltip=tooltip, style_function=lambda x: {'fillColor': 'transparent','color':'transparent'}).add_to(m)

out_html = 'figures/choropleth_kreis_interactive.html'
m.save(out_html)
print('Interaktive Karte gespeichert als', out_html)
m
```
