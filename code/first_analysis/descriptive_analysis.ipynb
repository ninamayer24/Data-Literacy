# Descriptive Analysis of Bridge Statistics in Sachsen-Anhalt and Thüringen
This script contains descriptive analysis of bridges in the German states Sachsen-Anhalt and Thüringen. 

```python

import pandas as pd
import matplotlib.pyplot as plt
```

```python
# load data
data = pd.read_csv('../../data/reduced_bridge_statistic_germany.csv', sep=';')
data.head()
```

```python
# only bridges that are in Sachsen-Anhalt or Thüringen
data = data[(data['Bundeslandname'] == 'Sachsen-Anhalt') | (data['Bundeslandname'] == 'Thueringen')]
data.head()
```

```python
# How many bridges are located in Thüringen and how many in Sachsen-Anhalt?
count_sa = data[data['Bundeslandname'] == 'Sachsen-Anhalt'].shape[0]
count_th = data[data['Bundeslandname'] == 'Thueringen'].shape[0]
print("Number of bridges in Sachsen-Anhalt:", count_sa)
print("Number of bridges in Thüringen:", count_th)
```

```python
# plot the distribution of Zustandsnote for Sachsen-Anhalt and Thüringen including means and standard deviations 
mean_sa = data[data['Bundeslandname'] == 'Sachsen-Anhalt']['Zustandsnote'].mean()
mean_th = data[data['Bundeslandname'] == 'Thueringen']['Zustandsnote'].mean()

sd_sa = data[data['Bundeslandname'] == 'Sachsen-Anhalt']['Zustandsnote'].std()
sd_th = data[data['Bundeslandname'] == 'Thueringen']['Zustandsnote'].std()


fig, ax = plt.subplots(figsize=(10, 6))
ax.set_title("Distribution of Zustandsnote in Sachsen-Anhalt and Thüringen")
ax.set_xlabel("Zustandsnote")
ax.set_ylabel("Count")

ax.bar(
    x = data[data['Bundeslandname'] == 'Thueringen']['Zustandsnote'].value_counts().sort_index().index,
    height = data[data['Bundeslandname'] == 'Thueringen']['Zustandsnote'].value_counts().sort_index().values,
    color='orange',
    width=0.05,
    label='Thüringen')

ax.bar(
    x = data[data['Bundeslandname'] == 'Sachsen-Anhalt']['Zustandsnote'].value_counts().sort_index().index + 0.05,
    height = data[data['Bundeslandname'] == 'Sachsen-Anhalt']['Zustandsnote'].value_counts().sort_index().values,
    color='blue',
    width=0.05,
    label='Sachsen-Anhalt')

ax.axvline(mean_sa, color='blue', linestyle='--', ms = 4, alpha= 1.0, label='Mean Sachsen-Anhalt')
ax.axvline(mean_th, color='orange', linestyle='--', ms = 4, alpha= 1.0, label='Mean Thüringen')

ax.plot([mean_sa - sd_sa, mean_sa + sd_sa], [300,300], ls = '--', marker='^', color='blue', label='Std Sachsen-Anhalt')
ax.plot([mean_th - sd_th, mean_th + sd_th], [250,250], ls = '--', marker='^', color='orange', label='Std Thüringen')

ax.legend()
plt.show()

print("Sachsen-Anhalt - Mean Zustandsnote:", mean_sa)
print("Thüringen - Mean Zustandsnote:", mean_th)

```

```python
# How many bridges are in a very good condition (Zustandsnote <= 1.4) (in percent)?
very_good_sa = data[(data['Bundeslandname'] == 'Sachsen-Anhalt') & (data['Zustandsnote'] <= 1.4)].shape[0]
very_good_th = data[(data['Bundeslandname'] == 'Thueringen') & (data['Zustandsnote'] <= 1.4)].shape[0]  
percent_very_good_sa = (very_good_sa / count_sa) * 100
percent_very_good_th = (very_good_th / count_th) * 100
print(f"Percentage of very good bridges in Sachsen-Anhalt: {percent_very_good_sa} %.")
print(f"Percentage of very good bridges in Thüringen: {percent_very_good_th} %.")

# How many bridges are in an insufficient condition (3,5 <= Zustandsnote <= 4.0) (in percent)?
insufficient_sa = data[(data['Bundeslandname'] == 'Sachsen-Anhalt') & (data['Zustandsnote'] >= 3.5) & (data['Zustandsnote'] <= 4.0)].shape[0]
insufficient_th = data[(data['Bundeslandname'] == 'Thueringen') & (data['Zustandsnote'] >= 3.5) & (data['Zustandsnote'] <= 4.0)].shape[0]  
percent_insufficient_sa = (insufficient_sa / count_sa) * 100
percent_insufficient_th = (insufficient_th / count_th) * 100
print(f"Percentage of insufficient bridges in Sachsen-Anhalt: {percent_insufficient_sa} %.")
print(f"Percentage of insufficient bridges in Thüringen: {percent_insufficient_th} %.")


# Out of the first quantile of bridges with the best condition, how many of them are in Sachsen-Anhalt and how many in Thüringen?
quantile_25 = data['Zustandsnote'].quantile(0.25)
print("25th percentile of Zustandsnote:", quantile_25)
best_bridges_sa = data[(data['Bundeslandname'] == 'Sachsen-Anhalt') & (data['Zustandsnote'] <= quantile_25)].shape[0]
best_bridges_th = data[(data['Bundeslandname'] == 'Thueringen') & (data['Zustandsnote'] <= quantile_25)].shape[0]
print("Number of bridges in the first quantile (best condition) in Sachsen-Anhalt:", best_bridges_sa)
print("Number of bridges in the first quantile (best condition) in Thüringen:", best_bridges_th)

```
